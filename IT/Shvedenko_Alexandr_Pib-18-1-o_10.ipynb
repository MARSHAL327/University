**LOCAL AND GLOBAL VARIABLES**

Various variables can be created and used within a function. To avoid problems, variables defined inside the function are not visible from the outside.

```python
f = 10 
def factorial(n):    
    f = 1   
    for i in range(2, n+1):     
        f = f * i   
    print("In the function, f =", f)    
    return f 
f = 10 
print(f) 
print(factorial(8)) 
print("Out of function") 
print(f)
```
