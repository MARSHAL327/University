**ALGORITHMS WITH LOOPS**

**Example of an algorithm using a loop**

**Performing initialization - setting the initial values of variables that we will need later**

**We write the values of the first two Fibonacci numbers;we write them to variables a and b and then store the two found Fibonacci numbers needed to find the next number.**

```python
#initialization(performed only once at the beginning of the algorithm)
a=1#first number
b=1#second number
```

**Then we write the code and go to the next number.We execute it several times,getting a new Fibonacci number**

```python
c=a+b
a=b
b=c
print(b)
```

```python
#initialization(executed only once at the beginning of the algorithm)
a=1
b=1
for i in range(10):
    #body loop:will be executed 10 times
    c=a+b
    a=b
    b=c
    print(i,c)
```

```python
a = 1 # first number 
b = 1 # second number
fib = []
# create enpty list fib 
for i in range(25):   
    c = a + b   
    a = b    
    b = c    
    fib.append(b) # wrote an element to the end of the fib list   
    
print(fib)
```

**LOCAL AND GLOBAL VARIABLES**

Various variables can be created and used within a function. To avoid problems, variables defined inside the function are not visible from the outside.

```python
f = 10 
def factorial(n):    
    f = 1   
    for i in range(2, n+1):     
        f = f * i   
    print("In the function, f =", f)    
    return f 
f = 10 
print(f) 
print(factorial(8)) 
print("Out of function") 
print(f)
```

**RETURNING TO THE FUNCTIONS**

The asterisk in the signature puts additional parentheses around the arguments.

```python
#the following two calls produce the same results
def test1(*args): 
    print(args) 
test1(1, 2, 3) 

def test2(args):   
    print(args) 
test2( (1,2,3) )
```

You can combine list variables and regular variables (provided that there is only one variable with an asterisk):

```python
def my_print(sep, *args):    
    for x in args:       
        print(x, end = sep)
```

```python
my_print('----', 7, 8, 9, 'hello')
```
